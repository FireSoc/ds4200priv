```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("data/Temperature_Anomaly.csv")
df["Year"] = pd.to_datetime(df["Year"], format="%Y")
```

```python
df
```

```python
alt.Chart(df).mark_circle().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N")
).properties(width=450)
```

```python
df_feb_aug = df[(df["Month"] == "Feb") | (df["Month"] == "Aug")]
df_feb_aug
```

```python
alt.Chart(df_feb_aug).mark_circle().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_square().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
    shape = alt.Shape("Month:N")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_text().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
    text = alt.Text("Month:N")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_tick().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
    text = alt.Text("Month:N")
).properties(width=450)
```

```python
color_domain = [-1, 0, 0.5, 1.8]
color_range = ["0000ff", "#88ccff", "#f98d5d", "#ff0000"]

alt.Chart(df_feb_aug).mark_tick(size=200, thickness=3).encode(
    x = alt.X("Year:T"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range))
).properties(width=400, height=200)
```

```python
alt.Chart(df_feb_aug).mark_line(point=False, interpolate="bundle").encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_trail(point=False, interpolate="bundle").encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
    size = alt.Size("Year:T")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_area(point=False, interpolate="basis").encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
    size = alt.Size("Year:T")
).properties(width=450)
```

```python
alt.Chart(df_feb_aug).mark_bar().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Month:N"),
).properties(width=450)
```

```python
data =  [{"category": "Rain", "value": 75, "order": 3},
      {"category": "Shade", "value": 10, "order": 1},
      {"category": "Sun", "value": 15, "order": 2}
    ]

df_pie = pd.DataFrame(data)
df_pie.head()
```

```python
alt.Chart(data=df_pie).mark_arc().encode(
    theta = alt.Theta("value:Q",), 
    color = alt.Color("category:N"), 
    order = "order:O"
)
```

```python
color_domain = ["Rain", "Shade", "Sun"]
color_range = ["blue", "#674028", "#DEAC58"]

alt.Chart(data=df_pie).mark_arc().encode(
    theta = alt.Theta("value:Q",), 
    color = alt.Color("category:N", scale = alt.Scale(domain=color_domain, range=color_range)), 
    order = "order:O"
)
```

```python
alt.Chart(data=df_pie).mark_bar().encode(
    x = alt.X("value:Q"),
    color = alt.Color("category:N", scale = alt.Scale(domain=color_domain, range=color_range)), 
    tooltip = "category:N"
).properties(height=250)
```
