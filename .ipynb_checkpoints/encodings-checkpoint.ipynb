```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("data/Temperature_Anomaly.csv")
df
```

```python
df["Year"] = pd.to_datetime(df["Year"], format="%Y")
```

```python
alt.Chart(df).mark_point().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
)
```

```python
alt.Chart(df).mark_point().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    size = alt.Size("Temperature_Anomaly:Q")
)
```

```python
alt.Chart(df).mark_point().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q")
)
```

```python
alt.Chart(df).mark_point().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(range="diverging"))
)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]

alt.Chart(df).mark_point().encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range))
)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]

alt.Chart(df).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range)),
    opacity = alt.Opacity("Temperature_Anomaly:Q")
)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]

alt.Chart(df).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range)),
    opacity = alt.value(0.5)
).properties(width=500)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]

alt.Chart(df).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range)),
    opacity = alt.value(0.5),
    shape = alt.Shape("Month:O")
).properties(width=500)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]

alt.Chart(df).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range)),
    opacity = alt.value(0.5),
    tooltip = ["Month:O", "Year:T"]
).properties(width=500)
```

```python
color_domain = [-1, 0, 1.4]
color_range = ["blue", "lightblue", "orange"]
order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

alt.Chart(df).mark_point(filled=True).encode(
    x = alt.X("Year:T"),
    y = alt.Y("Temperature_Anomaly:Q"),
    color = alt.Color("Temperature_Anomaly:Q", scale = alt.Scale(domain=color_domain, range=color_range)),
    opacity = alt.value(0.5),
    tooltip = ["Temperature_Anomaly:Q"],
    column = alt.Column("Month:O", sort=order)
).properties(width=200, height=150)
```
